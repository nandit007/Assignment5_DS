```python
#Create a function which will take a list as an argument and return the product of all the numbersafter creating a flat list.Use the below-given list as an argument for your function.
list1 = [1,2,3,4, [44,55,66, True], False, (34,56,78,89,34), {1,2,3,3,2,1}, {1:34, "key2": [55, 67, 78, 89], 4: (45,
22, 61, 34)}, [56, 'data science'], 'Machine Learning']
import functools
def prd(l):
    l1=[]
    for i in l:
        if type(i)==int:
            l1.append(i)
        if type(i)==list:
            for j in i:
                if type(j)==int:
                    l1.append(j)
        if type(i)==tuple:
            for j in i:
                if type(j)==int:
                    l1.append(j)
        if type(i)==set:
            for j in i:
                if type(j)==int:
                    l1.append(j)   
        if type(i)==dict:
            a=list(i.keys())
            b=list(i.values())
            for j in a:
                if type(j)==int:
                    l1.append(j)
            for j in b:
                if type(j)==int:
                    l1.append(j) 
                if type(j)==list:
                    for m in j:
                        if type(m)==int:
                            l1.append(m)
                    
                    
    l2=functools.reduce(lambda x,y:x*y,l1)
    return l1,l2



prd(list1)

            
            
            
            
            
            
            
            
            
            
            
           
                                  
            
            
                    
                    
                      
```
